```python
X = trn[trn['class'] != 0][['r']].values
y = trn[trn['class'] != 0]['dered_r'].values
print(X.shape, y.shape)
```

# RMSE 손실함수 정의

```python
rmse = lambda y, p: np.sqrt(np.mean((y - p) ** 2))
```

## NumPy를 이용한 선형회귀# 

```python
alpha = .001
theta0 = .1
theta1 = .1
epoch = 4

x = X.flatten()
print(x.shape)
```

```python
rmses = []
for i in tqdm(range(epoch)):
    for x_i, y_i in zip(x, y):
        p_i =  theta0 + theta1 * x_i
        theta0 = theta0 + alpha * (y_i - p_i)
        theta1 = theta1 + alpha * (y_i - p_i) * x_i
    rmses.append(rmse(y, theta0 + theta1 * x))
```

```python
plt.plot(rmses)
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(X, y, alpha=.1)
plt.plot(X, theta0 + theta1 * x, color='darkorange', linewidth=.5)
plt.xlim(0, 40)
plt.ylim(0, 40)
plt.xlabel('g')
plt.ylabel('dered_g')
```
