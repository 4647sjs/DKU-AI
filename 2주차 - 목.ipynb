```python
data_dir = Path('./input/')
trn_file = data_dir / 'train.csv'
seed = 42
trn = pd.read_csv(trn_file, index_col=0)
print(trn.shape)
trn.head()
```

```python
X = trn[trn['class'] != 0][['r']].values
y = trn[trn['class'] != 0]['dered_r'].values
print(X.shape, y.shape)
```

# RMSE  

```python
rmse = lambda y, p: np.sqrt(np.mean((y - p) ** 2))
```

## NumPy  # 

```python
alpha = .001
theta0 = .1
theta1 = .1
epoch = 4

x = X.flatten()
print(x.shape)
```

```python
rmses = []
for i in tqdm(range(epoch)):
    for x_i, y_i in zip(x, y):
        p_i =  theta0 + theta1 * x_i
        theta0 = theta0 + alpha * (y_i - p_i)
        theta1 = theta1 + alpha * (y_i - p_i) * x_i
    rmses.append(rmse(y, theta0 + theta1 * x))
```

```python
plt.plot(rmses)
```
